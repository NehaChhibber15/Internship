Q1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url=https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
a) Rank
b) Name
c) Artist
d) Upload date
e) Views

```python
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import ElementClickInterceptedException
import warnings
warnings.filterwarnings('ignore')
```

```python
driver= webdriver.Chrome(r"C:/Users/Suncity/Downloads/chromedriver_win32/chromedriver.exe")
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
#creating empty
Rank=[]
Name=[]
Artist=[]
Upload_date=[]
Views=[]
#scraping rank
rank=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
for i in rank:
    if i.text is None:
        Rank.append("--")
    else:
        Rank.append(i.text)
time.sleep(4)
#scraping name
name=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
for i in name:
    if i.text is None:
        Name.append("--")
    else:
        Name.append(i.text)
time.sleep(4)
artist=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
for i in artist:
    if i.text is None:
        Artist.append("--")
    else:
        Artist.append(i.text)
time.sleep(4)
upload_date=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
for i in upload_date:
    if i.text is None:
        Upload_date.append("--")
    else:
        Upload_date.append(i.text)
time.sleep(4)
view=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
for i in view:
    if i.text is None:
        Views.append("--")
    else:
        Views.append(i.text)
```

```python
df=pd.DataFrame({'Rank':Rank,'Name':Name,'Artist':Artist,'Upload_date':Upload_date,'Views(in billions)':Views})
df
```

```python
#creating empty list
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]


# fetching Data_type
data_type = driver.find_elements_by_xpath('//table[@cellpadding="5"][1]/tbody/tr/td[2]/p')
for i in data_type:
    if i.text is None :
        Data_type.append("--") 
    else:
        Data_type.append(i.text)
        
time.sleep(3)

#fetching Task
task = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]')
for i in task:
    if i.text is None :
        Task.append("--") 
    else:
        Task.append(i.text)
        
time.sleep(3)  
#fetching Attribute_type
attribute_type = driver.find_elements_by_xpath('//table[@cellpadding="5"][1]/tbody/tr/td[4]/p')
for i in attribute_type:
    if i.text is None :
        Attribute_type.append("--") 
    else:
        Attribute_type.append(i.text)
        
time.sleep(3)  
#fetching No_of_instances
no_of_instances = driver.find_elements_by_xpath('//table[@cellpadding="5"][1]/tbody/tr/td[5]/p')
for i in no_of_instances: 
    if i.text is None :
        No_of_instances.append("--") 
    else:
        No_of_instances.append(i.text)
        
time.sleep(3)  
#fetching No_of_attribute
no_of_attribute= driver.find_elements_by_xpath('//table[@cellpadding="5"][1]/tbody/tr/td[6]/p')
for i in no_of_attribute:
    if i.text is None :
        No_of_attribute.append("--") 
    else:
        No_of_attribute.append(i.text)
time.sleep(3)
#fetching Year
year= driver.find_elements_by_xpath('//table[@cellpadding="5"][1]/tbody/tr/td[7]/p')
for i in year:
    if i.text is None :
        Year.append("--") 
    else:
        Year.append(i.text)

```

```python
# fetching Dataset_name
Dataset_name=[]
dataset_name = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]")
for i in dataset_name:
    if i.text is None :
        Dataset_name.append("--") 
    else:
        Dataset_name.append(i.text)
        
```

```python
#Combining all the lists into a single dataframe
df=pd.DataFrame({'Dataset_name':Dataset_name,'Data_type': Data_type, 
               'Task': Task,
                'Attribute_type': Attribute_type,
                'No_of_instances': No_of_instances,
                'No_of_attribute':No_of_attribute,'Year':Year})
                

                
df
```
